```python
import pandas as pd
df = pd.read_csv(r"C:\Users\LENOVO\Downloads\.csv file\messy_retail_sales_dataset.csv")
```

```python
df
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df
```

```python
print(df.head())
```

```python
#Total Revenue & Profit
```

```python
total = df[['Revenue', 'Profit']].sum()
print("Total Revenue & Profit")
print(total)
```

```python
total.to_csv("total_revenue_profit.csv", header = True)
```

```python
#Revenue by Region
```

```python
revenue_region = df.groupby('Region')['Revenue'].sum().sort_values(ascending = False).head(10)
print("Revenue by Region")
print(revenue_region)
```

```python
revenue_region.to_csv("revenue by region.csv")
```

```python
#Top products by Revenue
```

```python
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending = False).head(10)
print("Top 10 Products by Revenue")
print(top_products)
```

```python
top_products.to_csv("top_products.csv")
```

```python
#Monthly Revenue Trend 
```

```python
monthly_revenue = df.groupby(df['Order_Date'].dt.to_period('M'))['Revenue'].sum()
print("Monthly Revenue Trend")
print(monthly_revenue)
```

```python
monthly_revenue.to_csv("montthly_revenue_trend.csv")
```

```python
#Profit analysis
```

```python
profit_analysis = df.groupby('Category')['Profit'].sum().sort_values(ascending = False)
print("Profit Analysis Breakdown")
print(profit_analysis)
```

```python
profit_analysis.to_csv("profit_analysis_breakdown.csv")
```
