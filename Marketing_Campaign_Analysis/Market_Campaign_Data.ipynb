```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.describe()
```

```python
df['CTR'] = df['Clicks'] / df['Impressions']
df['Conversion_Rate'] = df['Conversions'] / df['Clicks']
df['ROI'] = df['ROI'] / df['Budget']

df[['CTR', 'Conversion_Rate', 'ROI']].describe()
```

```python
# Revenue by Channel
```

```python
Revenue_by_Channel = df.groupby('Channel')['Revenue'].sum()
```

```python
plt.figure()
Revenue_by_Channel.plot(kind = 'bar')
plt.title("Total Revenue by Channel")
plt.xlabel("Channel")
plt.ylabel("Revenue")
plt.xticks(rotation = 30)
plt.show()
```

```python
#  Revenue by Channel
```

```python
ROI_by_Channel = df.groupby('Channel')['ROI'].mean()
```

```python
plt.figure()
Revenue_by_Channel.plot(kind = 'bar')
plt.title("Average ROI by Channel")
plt.xlabel("Channel")
plt.ylabel("Average ROI")
plt.xticks(rotation = 30)
plt.show()
```

```python
plt.figure()
plt.scatter( df['Budget'],  df['Revenue'] )
plt.title("Budget vs Revenue")
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.show()
```
